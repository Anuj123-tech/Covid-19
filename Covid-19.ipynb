# Data Visualization

```python
sns.regplot(x='Confirmed',y='Recovered',data=df,marker='*',color='blue',fit_reg=False)
```

```python
sns.regplot(x='Confirmed',y='Deaths',data=df,marker="+",color='red',fit_reg=False)
```

```python
plt.figure(figsize=(14,7))
plt.style.use('seaborn-darkgrid')
# Group variable using "hue" that will produce lines with different colors
sns.lineplot(x="Confirmed" , y="Deaths", data=df)
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(14,7))
plt.style.use('seaborn-darkgrid')
# Group variable using "hue" that will produce lines with different colors
sns.lineplot(x="Confirmed" , y="Deaths / 100 Recovered", data=df)
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.scatterplot(x='New recovered' , y='New deaths' ,data=df,color='green')
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.scatterplot(x='New recovered' , y='New deaths' , hue='Active',data=df )
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.set(rc={"axes.facecolor":"#283747", "axes.grid":False,'xtick.labelsize':14,'ytick.labelsize':14})
sns.lineplot(x=df['Deaths / 100 Cases'],y=df['Deaths / 100 Recovered'],data=df , color = "#FF5722" , linewidth = 2 )
sns.lineplot(x=df['Deaths / 100 Cases'],y=np.cos(df['Deaths / 100 Recovered']),data=df , color = "#FFEB3B" , linewidth = 2)
plt.show()
```

```python
plt.figure(figsize=(20,8)) 
sns.lineplot(data=df[['Confirmed', 'Deaths', 'Recovered', 'Active', 'New cases','New deaths']]) 
plt.show()
```

```python
plt.figure(figsize=(8,10)) 
sns.barplot(x=df['Confirmed'], y=df['Recovered']) 
plt.show()
```
